```python
BOARD = []
dirX = [0,1] #2*2의 좌상단 블록 기준으로 찾을 것임.
dirY = [0,1]

def clearBlock(c,r):
    BOARD[r] = BOARD[r][:c] + "##" + BOARD[r][c+2:]#같은것을 찾았으면 #으로 바꿔줌.
    BOARD[r+1] = BOARD[r+1][:c] + "##" + BOARD[r+1][c+2:]
    print("current BOARD: ",BOARD)
    return

def findTwoByTwo(c,r):
    ret = BOARD[r][c].isalpha()
    for x in dirX:
        for y in dirY:
            ret = ret and (BOARD[r][c] == BOARD[r+y][c+x])
    return ret

def findAllTwoByTwo(m, n):
    point = []
    for col in range(m-1):
        for row in range(n-1):
            if findTwoByTwo(col, row):
                point.append((col, row))
    print("point result",point) #point result 안에 tuple.
    return point

def downBlock(m):
    for vertical in range(len(BOARD)):
        BOARD[vertical] = BOARD[vertical].replace('#', '')#현재 것을 공란으로 대체.
        BOARD[vertical] = "#"*(m-len(BOARD[vertical])) + BOARD[vertical]
        #지운것은 카운트가 안되니까 전체 길이에서 현재 없어진 블록과의 길이 차를 구한 뒤,
        #그 길이만큼 #을 채워준다.

def findRemovedBlock(m, n):
    cnt = 0
    for vertical in BOARD:
        for c in vertical:
            if c == '#':
                cnt += 1
    return cnt

def solution(m, n, board):

    for row in range(n):
        string = ""
        for col in range(m):
            string += board[col][row]
        BOARD.append(string)
    print("Initial BOARD : ",BOARD)

    while(True):
        points = findAllTwoByTwo(m, n)
        if points == []: #같은게 없으면 탈출.
            break

        for point in points:
            clearBlock(point[0], point[1])#clear

        downBlock(m)
        print("After putting down blocks: ",BOARD)

    answer = findRemovedBlock(m, n)
    return answer
```
